```python
exon_mouse['gbid'].iloc[0]
```

```python
from Bio import Entrez
from Bio import SeqIO
from Bio.SeqUtils import GC, molecular_weight

Entrez.email = 'Hello@example.com'
handle = Entrez.efetch(db = 'nucleotide', id = exon_mouse["gbid"].iloc[0], rettype = 'gb')
seq_record = SeqIO.read(handle, 'gb')
print(molecular_weight(seq_record.seq))
handle.close()
```

```python
exon_mouse10 = exon_mouse.iloc[0:10,:]
```

```python
exon_mouse10.reset_index(drop=True, inplace=True)
```

```python
mws = []
for gbid in exon_mouse10["gbid"]:
    handle = Entrez.efetch(db = 'nucleotide', id = gbid, rettype = 'gb')
    seq_record = SeqIO.read(handle, 'gb')
    mws.append(molecular_weight(seq_record.seq))
    handle.close()

exon_mouse10.insert(len(exon_mouse10.columns), 'mw', mws)
```

```python
exon_mouse10
```

**NOTE** We have to filter on exons to be able to query the Entrez nucleotide database otherwise there will be gbid for other types and you will need to specify another db value e.g. protein.
